```python
# New test file
import numpy as np
import pandas as pd
```

```python
df = {'x':['bar', 'bar', 'bar', 'foo', 'foo', 'foo']
, 'y':['a','b','c','a','b','c']
, 'z':[2, 3, 5, 7, 1, 2]}

df = pd.DataFrame(df)
df
```

## Grouping data

```python
# groupby returns a pandas series test 01
# group by x and values from z then applying np.sum function 
s = df.groupby('x')['z'].aggregate(np.sum)
df1 = pd.DataFrame(s)
df1
```

```python
df3 = pd.DataFrame(df.groupby('x')['z'].aggregate(np.cumsum))
df3
```

```python
df4 = pd.DataFrame(df.set_index('y').groupby('x')['z'].transform(np.cumsum))
df4
```

## Transfoming dataframe

```python
df2 = pd.DataFrame(df.groupby('x')['z'].transform(np.cumsum))
df.loc[:, 'z_cumsum'] = df.groupby('x')['z'].transform(np.cumsum)
df.loc[:, 'z_ave'] = df.groupby('x')['z'].transform(np.mean)
df.loc[:, 'z_std'] = df.groupby('x')['z'].transform(np.std)
df.loc[:, 'z_score'] = (df['z'] - df['z_ave'])/df['z_std']
df
```

```python
df_a = pd.DataFrame(dict(A=list('aabb'), B=[1, 2, 3, 4], C=[0, 9, 0, 9]))
# df.set_index('A').groupby(level='A').transform('mean') is the same as the below one
df_a.set_index('A').groupby('A').transform('mean')
```

## Restructuring DataFrame

```python
# Can just slicing the former df by using loc 
df = df.loc[:, ['x', 'y', 'z']]
df.info()
df
```

## Pivoting

```python
# generating the main entries with conrresponding values
df5 = df.pivot(index='x', columns='y', values='z')
df5
```

## Melt

```python
df6 = df5.reset_index().melt(id_vars=df5.index.name, value_vars=['a','b','c'], value_name='z')
df6
```

## Merge

```python
df_a = pd.DataFrame({'x':['a', 'b', 'c', 'd'], 'y':[1,2,3,4]})
df_b = pd.DataFrame({'x':['a', 'b', 'c', 'd'], 'z':[2,1,3,np.nan]})
df_a
```

```python
df_b
```

```python
# meger by columns that two dataframes have commomnly using inner
df_m1 = pd.merge(df_a, df_b, how='inner', on='x')
df_m1
```

```python
df_m1.loc[:, ['x','y']]
```

```python
# inplace index means we dont want to see a extra column of numerical index
df_a.set_index('x', inplace=True)
df_a
```

```python
df_b.set_index('x', inplace=True)
df_b
```

```python
# merge by index when two dataframes have the same indices and orders
df_m2 = pd.merge(df_a, df_b, left_index=True, right_index=True)
df_m2
```

```python
df_m1.set_index('x')
```

```python
df_a = pd.DataFrame({'x':['a', 'b', 'c', 'd'], 'y':[1,2,3,4]})
df_b = pd.DataFrame({'x':['a', 'b', 'c', 'd'], 'z':[2,1,3,np.nan]})

df_a.set_index('x', inplace=True)
df_b.set_index('x', inplace=True)
# drop a nan value by index
df_b.drop('d', axis=0, inplace=True) 
df_a
```

```python
df_b
```

```python
# 按照右边边的df的index来合并,
df_m3 = pd.merge(df_a, df_b, on='x', how='right')
df_m3
```

```python
# 按照左边的df的index来合并
df_m4 = pd.merge(df_a, df_b, on='x', how='left')
df_m4
```

## Concatenate

```python
df_a = pd.DataFrame({'x':['a', 'b', 'c', 'd'], 'y':[1,2,3,4]})
df_b = pd.DataFrame({'x':['a', 'b', 'c', 'd'], 'y':[2,1,3,np.nan]})
```

```python
df_a
```

```python
df_b
```

```python
# this result will produce dupilcated indices 
df_1 = pd.concat([df_a, df_b])
df_1
```

```python
# to fix the duplicated index using ignore_index=True
df_1 = pd.concat([df_a, df_b], ignore_index=True)
df_1
```
